# Understanding Word Embeddings: The Foundation of Language Models

In this post, we'll explore one of the key concepts that powers large language models (LLMs) - word embeddings. Don't worry if it sounds complex right now, because by the end of this post, you'll have a clear idea of what word embeddings are and how they transform plain text into a form that computers can understand and learn from.

## What Are Word Embeddings?
When we read or speak, we understand words as they relate to one another based on meaning. However, computers don't work this way. To get computers to process and make sense of text, we need to convert words into numbers. That's where word embeddings come in. Word embeddings transform words into vectors, which are simply lists of numbers that hold meaningful relationships between words.
Think of a word as a point in space. The closer two words are, the more similar they are in meaning. For example, "cat" and "dog" would have vectors that are close to each other, but "cat" and "car" would be far apart.

## Why Do We Need Word Embeddings?
Computers can only handle numbers, not text. When you're training a model to understand or generate text (like ChatGPT), it needs to work with numbers instead of raw text. So, we first embed words into numerical form before training. This is crucial for training a neural network model, which powers LLMs like GPT.

## Creating Word Embeddings: Step-by-Step
Let's go step-by-step and build the foundation for converting words into embeddings using Python.

1. Start With Text
We'll start with a simple sentence. Imagine you have this text that you want the model to understand:

```python
sentence = "I love learning about machine learning"
```

## 2. Tokenize the Sentence

Before we create word embeddings, we first need to split this sentence into individual words. This process is called tokenization. In this example, each word will become a token.

```python
tokens = sentence.split()
print(tokens)
```

Here, we split the sentence into words based on spaces, and each word is now a token.

## 3. Assign a Unique ID to Each Word

Next, we need to give each unique word a number (because computers work with numbers). This is called indexing.

```python
vocab = {word: i for i, word in enumerate(set(tokens))}
print(vocab)
```

Each word in the sentence is now assigned a unique index. For example, "about" is represented by 0, and "learning" by 2.

## 4. Turn Words into Vectors (Word Embeddings)

Now, we'll create embeddings for these words. Instead of just assigning numbers to words, we'll give each word a vector of numbers. Each word will now be represented by a list of numbers that capture its meaning. For simplicity, let's make our embeddings random for now (we'll learn how to create meaningful embeddings in later posts).

```python
import numpy as np

# Generate random word embeddings (3-dimensional vector for each word)
embedding_dim = 3
word_embeddings = {word: np.random.randn(embedding_dim) for word in vocab}
print(word_embeddings)
```

Now, each word is represented as a vector of 3 numbers. These numbers don't mean much for now because they're random, but in real-world models, the embeddings will be learned from data to capture relationships between words.

## 5. Use the Embeddings

Finally, we can use these vectors (embeddings) in our models. For example, if you want to get the embedding for the word "learning", you can just look it up:



```python
print("Embedding for 'learning':", word_embeddings['learning'])
```

## Full Sample Code

Here's the complete code that shows how we tokenize a sentence, assign IDs to words, and create random word embeddings:

```python
import numpy as np

# Step 1: Define the sentence
sentence = "I love learning about machine learning"

# Step 2: Tokenize the sentence into individual words (tokens)
tokens = sentence.split()

# Step 3: Create a vocabulary with unique IDs for each word
vocab = {word: i for i, word in enumerate(set(tokens))}

# Step 4: Create random word embeddings for each word
embedding_dim = 3  # Dimensionality of the embeddings (can be higher in real applications)
word_embeddings = {word: np.random.randn(embedding_dim) for word in vocab}

# Step 5: Access the embedding for a specific word (e.g., 'learning')
print("Tokens:", tokens)
print("Vocabulary:", vocab)
print("Embedding for 'learning':", word_embeddings['learning'])
```

# Key Takeaways:

Word embeddings turn words into numbers so computers can understand them.
Each word is given a unique vector that holds meaning about the word's relationship to others.
In real-world models, these embeddings are learned and updated during training, helping the model better understand language.



```python
import matplotlib.pyplot as plt

# Create a figure
fig, ax = plt.subplots(figsize=(10, 10))

# Define the flowchart elements with examples
elements = [
    ("Input Text", "I love machine learning", (0.3, 0.9)),
    ("Tokenization", "Tokens: ['I', 'love', 'machine', 'learning']", (0.3, 0.7)),
    ("Vocabulary Creation", "{'I': 0, 'love': 1, 'machine': 2, 'learning': 3}", (0.3, 0.5)),
    ("Embedding Creation", 
     "'I' -> [0.5, 0.1, -0.3]\n'love' -> [0.4, 0.6, 0.2]\n'machine' -> [0.2, -0.4, 0.5]\n'learning' -> [-0.2, 0.3, 0.1]", 
     (0.3, 0.3)),
    ("Word Vector Representation", 
     "Similar words cluster together in vector space", (0.3, 0.1))
]

# Define rectangle properties
rect_width = 0.4
rect_height = 0.1

# Add boxes and examples to the diagram
for (title, example, position) in elements:
    # Draw rectangle
    rect = plt.Rectangle((position[0] - rect_width / 2, position[1] - rect_height / 2), 
                         rect_width, rect_height, edgecolor='black', facecolor='lightblue')
    ax.add_patch(rect)
    
    ax.text(position[0], position[1], title, ha='center', va='center', 
            fontsize=12)
    ax.text(position[0] + rect_width / 2 + 0.05, position[1], example, 
            ha='left', va='center', fontsize=10)

# Draw arrows between each step
for i in range(len(elements) - 1):
    start = elements[i][2]
    end = elements[i + 1][2]
    # Calculate arrow start and end positions (middle of bottom to middle of top)
    arrow_start = (start[0], start[1] - rect_height / 2)
    arrow_end = (end[0], end[1] + rect_height / 2)
    
    ax.annotate("", xy=arrow_end, xycoords='data', xytext=arrow_start, textcoords='data',
                arrowprops=dict(arrowstyle="->", lw=1.5, color='black'))

# Set limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')

# Title
plt.title("Process of Creating Word Embeddings with Examples", fontsize=16)

# Show the plot
plt.show()
```
